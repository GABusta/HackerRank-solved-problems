```python
import sys

def main(lista, A, B):
  union1 = set(lista).intersection(set(A))
  union2 = set(lista).intersection(set(B))

  salida = int(len(union1) - len(union2))
  del A, B

  for i in range(len(union1)):
    condicion = lista.count(list(union1)[i])
    if condicion > 0:
      salida += int(condicion) - 1
  for i in range(len(union2)):
    condicion = lista.count(list(union2)[i])
    if condicion > 0:
      salida -= int(condicion) - 1

  return salida
# -----------------------------------
sys.stdin = open('datos2.txt', 'r')

data = sys.stdin.readline().split()
n, m = int(data[0]), int(data[1])

# ----- Array 1 -----
data2 = sys.stdin.readline().split()
lista = [int(i) for i in data2]
# ----- Array A -----
data3 = sys.stdin.readline().split()
A = [int(i) for i in data3]
# ----- Array B -----
data4 = sys.stdin.readline().split()
B = [int(i) for i in data4]

# ----- Salida -----
del data, data2, data3, data4
salida = main(lista, A, B)
print(salida)
```
