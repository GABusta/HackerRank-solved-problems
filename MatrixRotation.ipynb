```python
import math
import os
import random
import re
import sys

#
# Complete the 'matrixRotation' function below.
#
# The function accepts following parameters:
#  1. 2D_INTEGER_ARRAY matrix
#  2. INTEGER r
#

def matrixRotation(matrix, r, m ,n):
  # Write your code here
  MatrixModif = matrix
  for j in range(r):
    ciclo = 1
    Amodif = list([[0 for m in range(n)] for i in range(m)])
    while (m - 2*ciclo >= 0) and (n - 2*ciclo >= 0):
      Amodif = OrdenadoColumnas(MatrixModif, Amodif, ciclo)
      if (n - 2*ciclo > 0):
        Amodif = OrdenadoFilas(MatrixModif, Amodif, ciclo)
      ciclo += 1
    MatrixModif = Amodif

  for i in range(m): 
    print(' '.join(map(str, MatrixModif[i][:])))
    
  return MatrixModif

# ----------------------------------------
def OrdenadoColumnas(matriz, a, ciclo):
  col1 = ciclo - 1
  col2 = n - ciclo
  fila1 = ciclo - 1 
  fila2 = m - ciclo

  for j in range(fila1, fila2 + 1):
    if j == fila1:
      a[j][col1] = matriz[j][col1 + 1]
      a[j][col2] = matriz[j + 1][col2]
    elif j == fila2:
      a[j][col1] = matriz[j - 1][col1]
      a[j][col2] = matriz[j][col2 - 1]
    else:
      a[j][col1] = matriz[j - 1][col1]
      a[j][col2] = matriz[j + 1][col2]

  return a
# ----------------------------------------
def OrdenadoFilas(matriz, a, ciclo):
  col1 = ciclo - 1
  col2 = n - ciclo
  fila1 = ciclo - 1 
  fila2 = m - ciclo

  for j in range(col1, col2 + 1):
    if j != col1 and j != col2: 
      a[fila1][j] = matriz[fila1][j + 1]
      a[fila2][j] = matriz[fila2][j - 1]

  return a
# ----------------------------------------

if __name__ == '__main__':
    first_multiple_input = input().rstrip().split()

    m = int(first_multiple_input[0])

    n = int(first_multiple_input[1])

    r = int(first_multiple_input[2])

    matrix = []

    for _ in range(m):
        matrix.append(list(map(int, input().rstrip().split())))

    matrixRotation(matrix, r, m ,n)
```
