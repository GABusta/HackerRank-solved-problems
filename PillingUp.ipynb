```python
import sys

def main(Ncasos, lista, cantidad):
  solucion = []
  for i in range(Ncasos):
    solucion.append('Yes')

  for i in range(Ncasos):
    ordenado = []
    # --- Ordenamiento datos
    listaRemocion = lista[i][:]
    for j in range(cantidad[i]):
      if listaRemocion[0] > listaRemocion[-1]:
        ordenado.append(listaRemocion[0])
        del listaRemocion[0]
      else:
        ordenado.append(listaRemocion[-1])
        del listaRemocion[-1]

    # --- si o no
    for j in range(cantidad[i]-1):
      if ordenado[j+1] > ordenado[j]:
        solucion[i] = 'No'

  return solucion
# -----------------------------------
sys.stdin = open('datos.txt', 'r')

data = sys.stdin.readline().split()
Ncasos = int(data[0])

lista, cantidad = [], []

for i in range(Ncasos*2):
  data2 = sys.stdin.readline().split()
  if (i+1) % 2 == 0:
    fila = []
    for j in range(len(data2)):
      fila.append(int(data2[j]))
    lista.append(fila)
  else:
    cantidad.append(int(data2[0]))

salida = main(Ncasos, lista, cantidad)
print(salida)
```
