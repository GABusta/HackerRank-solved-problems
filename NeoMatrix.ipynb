# **Problema HackerRank**

Neo has a complex matrix script. The matrix script is a  X  grid of strings. It consists of alphanumeric characters, spaces and symbols (!,@,#,$,%,&).

To decode the script, Neo needs to read each column and select only the alphanumeric characters and connect them. Neo reads the column from top to bottom and starts reading from the leftmost column.

If there are symbols or spaces between two alphanumeric characters of the decoded script, then Neo replaces them with a single space '' for better readability.

Neo feels that there is no need to use 'if' conditions for decoding.

Alphanumeric characters consist of: [A-Z, a-z, and 0-9].


**Input Format**

The first line contains space-separated integers  (rows) and  (columns) respectively.
The next  lines contain the row elements of the matrix script.

**Constraints**

Note: A  score will be awarded for using 'if' conditions in your code.

Output Format

Print the decoded matrix script.

Sample Output 0

*This is Matrix#  %!*

# Secci√≥n nueva

```python
#!/bin/python3

import math
import os
import random
import re
import sys


first_multiple_input = input().rstrip().split()

n = int(first_multiple_input[0])

m = int(first_multiple_input[1])

matrix = []

for _ in range(n):
    matrix_item = input()
    matrix.append(matrix_item)
```

```python
print(matrix)
```

```python
salida, arreglo = '', ''
va, vb = False, False
comp1, comp2, comp3 = re.compile('([A-Z])'), re.compile('([a-z])'), re.compile('([0-9])')
inicio = bool(re.search(comp1, matrix[0][0]) or re.search(comp2, matrix[0][0]) or re.search(comp3, matrix[0][0]))

for i in range(m):
  for j in range(n):
    data = matrix[j][i]
    # comp1, comp2, comp3 = re.compile('([A-Z])'), re.compile('([a-z])'), re.compile('([0-9])')
    resul = bool(re.search(comp1, data) or re.search(comp2, data) or re.search(comp3, data))
    # inicio = bool(re.search(comp1, matrix[0][0]) or re.search(comp2, matrix[0][0]) or re.search(comp3, matrix[0][0]))
    while resul == False:
      va, vb = False, True
      arreglo += data  
      break
    while resul == True and inicio == False and va == False and vb == True:
      salida += arreglo
      va, vb = True, False
      break  
    while resul == True and va == False and vb == True:
      salida += ' ' 
      break
    while resul == True:
      salida += data
      arreglo = ''
      va, vb = True, False
      inicio = True
      break
    while resul == False and i == m - 1 and j == n - 1:
      salida += arreglo
      break
print(salida)
```
